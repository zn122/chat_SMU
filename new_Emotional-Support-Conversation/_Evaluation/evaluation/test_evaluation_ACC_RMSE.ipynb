```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv('estimation_all2.csv', encoding='cp949')
test
```

```python
def top_k_accuracy(true_labels, predicted_labels, k=1):
    # 예측 레이블에서 상위 k개의 클래스를 가져옵니다.
    top_k_preds = predicted_labels[:, :k]

    # 정답 레이블이 상위 k개의 예측 레이블 중 하나에 포함되는지 확인합니다.
    correct_predictions = np.array([true_label in top_k for true_label, top_k in zip(true_labels, top_k_preds)])

    # 정확한 예측의 비율을 계산합니다.
    top_k_acc = np.mean(correct_predictions)

    return top_k_acc
```

```python
# top-3 emotion accuracy
true_labels = np.array(test['emotion'].tolist())
predicted_labels = np.array(test[['Emotion1','Emotion2','Emotion3']].values.tolist())
k = 3
accuracy = top_k_accuracy(true_labels, predicted_labels, k)
print(f"Top-{k} 정확도: {accuracy}")
```

```python
# top-2 emotion accuracy
true_labels = np.array(test['emotion'].tolist())
predicted_labels = np.array(test[['Emotion1','Emotion2']].values.tolist())
k = 2
accuracy = top_k_accuracy(true_labels, predicted_labels, k)
print(f"Top-{k} 정확도: {accuracy}")
```
