function ClickConnect() {
    console.log("Working");
    const button = document.querySelector("colab-toolbar-button");
    if (button) {
        button.click();
        console.log("Button clicked!");
    } else {
        console.log("Button not found. Retrying in 30 seconds...");
    }
}

// 최초 실행
ClickConnect();

setInterval(ClickConnect, 600000);
