```python
import json
import random
import csv
import copy

with open('/content/drive/MyDrive/아연/분포/0909/sentiment_chatbot_dataset_emotion_type_init_final.csv', newline='') as f:
    reader = csv.reader(f)
    tag_data = list(reader)

tag = copy.deepcopy(tag_data[1:])  # remove column name

```

```python
tag[1][2]#sentence
tag[1][3]#emotion
tag[1][4]#대표감정
tag[1][5]#init-final
tag[1][6]#최종 intensity
```

```python
import numpy as np
a = np.full(len(tag),111)
print(a)
tag = np.c_[tag, a]
tag[1][6]
```

```python
#fear
from transformers import AutoModelForSequenceClassification, AutoTokenizer
model = AutoModelForSequenceClassification.from_pretrained("garrettbaber/twitter-roberta-base-fear-intensity")
tokenizer = AutoTokenizer.from_pretrained("garrettbaber/twitter-roberta-base-fear-intensity")

for i in range(len(tag)):
  if (tag[i][4] == 'fear') or (tag[i][4] == 'nervousness') or (tag[i][4] == 'anxiety'):
    print(tag[i][2])
    print(tag[i][3])
    inputs = tokenizer(tag[i][2], return_tensors="pt")
    outputs = model(**inputs)
    tag[i][6] = float(outputs.logits)*5
    #tag[i][6] = round(float(outputs.logits)*5,0)
    print(tag[i][6])
```

```python
#sadness
from transformers import AutoModelForSequenceClassification, AutoTokenizer
model = AutoModelForSequenceClassification.from_pretrained("garrettbaber/twitter-roberta-base-sadness-intensity")
tokenizer = AutoTokenizer.from_pretrained("garrettbaber/twitter-roberta-base-sadness-intensity")

for i in range(len(tag)):
  if (tag[i][4] == 'sadness') or (tag[i][4] == 'shame') or (tag[i][4] == 'depression') or (tag[i][4] == 'guilt') or (tag[i][4] == 'pain'):
    print(tag[i][2])
    print(tag[i][3])
    inputs = tokenizer(tag[i][2], return_tensors="pt")
    outputs = model(**inputs)
    tag[i][6] = float(outputs.logits)*5
    #tag[i][6] = round(float(outputs.logits)*5,0)
    print(tag[i][6])
```
